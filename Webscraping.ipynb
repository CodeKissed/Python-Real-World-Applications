```python
import requests
from bs4 import BeautifulSoup

loadpage= requests.get("https://www.gumtree.com.au/s-property-for-rent/melbourne/apartment+rent/k0c18364l3001317", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})

pagecon = loadpage.content

soup=BeautifulSoup(pagecon,"html.parser")
```

```python
#all=soup.find_all("div",{"class":"user-ad-row__details"})
#all[0].find("p",{"class":"user-ad-row__title"}).text
page=soup.find_all("a",{"class":"page-number-navigation__link"})
lastpage=len(page)


```

```python
gumtree=[]
base_url="https://www.gumtree.com.au/s-property-for-rent/melbourne/apartment+rent/page-"
base_url2="/k0c18364l3001317"
for page in range(0,lastpage,1):
    print(base_url+str(page)+base_url2)
    callpage=requests.get(base_url+str(page)+base_url2)
    savepage=callpage.content
    soup=BeautifulSoup(savepage,"html.parser")
    all=soup.find_all("div",{"class":"user-ad-row__details"})                 
                   
    for item in all:
        gt={}
        gt["Headline"]=item.find("p",{"class":"user-ad-row__title"}).text
        try:
            gt["Price"]=item.find("span",{"class":"user-ad-price__price"}).text
        except:
            gt["Price"]=None
        try:
            gt["Bathroom"]=item.find_all("li",{"class":"user-ad-attributes__attribute"})[0].text
        except:
            gt["Bathroom"]=None
        try:
            gt["Bedroom"]=item.find_all("li",{"class":"user-ad-attributes__attribute"})[1].text
        except:
            gt["Bedroom"]=None
        try:
            gt["Property Type"]=item.find_all("li",{"class":"user-ad-attributes__attribute"})[2].text
        except:
            gt["Property Type"]=None
        gt["Location"]=item.find("span",{"class":"user-ad-row__location-area"}).text
        gumtree.append(gt)

```

```python
import pandas
df=pandas.DataFrame(gumtree)
df
```

```python
df.to_csv("Gumtree.csv")
```
